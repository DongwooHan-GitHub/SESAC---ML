```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc("font", family = "gulim")
plt.rc("axes", unicode_minus = False)
```

```python
df1 = pd.read_excel("./dataset/stock price.xlsx")
df2 = pd.read_excel("./dataset/stock valuation.xlsx")

print("df1 = \n", df1.head())
print("df2 = \n", df1.head())
```

```python
## concat > 붙히는 작업
pd.concat([df1, df2], join= "outer", ignore_index=True)
pd.concat([df1, df2], axis=1, join= "outer", ignore_index=True)
```

```python
## join
df1.join(df2, lsuffix="x", rsuffix= "y", how= "inner")
df2.join(df1, lsuffix="x", rsuffix= "y", how= "inner")
```

```python
## merge > 병합
# pd.merge(df1, df2, on = "id", how = "left")
pd.merge(df1, df2, on = "id", how = "left")
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic_group = titanic.groupby("survived")

titanic_group[["age", "fare"]].mean()

```

```python
titanic["age"].agg(["mean","median","std","min","max"])
titanic_group["age"].agg(["mean","median","std","min","max"])
```

```python
sns.regplot(titanic,
            x= "age",
            y = "fare",
            fit_reg = True,
            color = "r",
            order = 1,
            label ="titanic")
plt.title("나이 vs 요금")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()
plt.show()
```

```python
penguins = sns.load_dataset("penguins")

sns.lmplot(penguins,
        x = "bill_length_mm",
        y = "bill_depth_mm",
        hue = "species",
        row = "sex",
        sharex= False,
        sharey= False)

plt.show()
```

```python
sns.scatterplot(penguins,
                x = "bill_length_mm",
                y = "bill_depth_mm",
                s = 100,
                hue = "species",
                style= "island",
                palette= "pastel")
plt.show()
```
