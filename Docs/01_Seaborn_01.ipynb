```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = "gulim")
plt.rc('axes', unicode_minus = False)
```

```python
titanic = sns.load_dataset("titanic")
print("titanic = \n", titanic.head())
print(titanic.info())
```

```python
## missing
print("missing value = ")
# print(titanic.isnull().sum())

titanic_df = titanic.loc[:, :"fare"]
titanic_df.info()

titanic_df.dropna(axis = 0, inplace= True)
titanic_df.info()
```

```python
##
titanic_df.head()
```

```python
## sns.regplot
fig, ax = plt.subplots(figsize = (5,5))
sns.regplot(titanic_df, x = "age", y= "fare")
plt.show()
# fig.savefig("titanic_reg.png")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#  Titanic   ( titanic_df   )
# titanic_df = sns.load_dataset("titanic")

# regplot 
sns.regplot(data=titanic_df,  # 'titanic_df'    
            x="age", y="fare",
            fit_reg=True,
            color="r",
            marker="*",
            order=1,
            label="titanic data",
            scatter_kws={"fc": "b", "ec": "b", "s": 100, "alpha": 0.3},
            line_kws={"lw": 2, "ls": "--", "alpha": 0.8}  #  
            )
plt.title(" VS ")
plt.xlabel("")
plt.ylabel("")
#  
plt.legend(loc=0)

#  
plt.show()
```

```python
# diamonds  
diamonds = sns.load_dataset("diamonds")
diamonds_df = diamonds[diamonds["cut"] == "Ideal"]

# regplot 
sns.regplot(data=diamonds_df.sample(n=1000),  # 1000  
            x="carat",
            y="price",
            order=2,
            scatter_kws={"fc": "b", "ec": "b", "s": 100, "alpha": 0.5},
            line_kws={"color": "red"}  # color= 
            )

#     
plt.title("Carat VS Price")
plt.xlabel("Carat")
plt.ylabel("Price")

#  
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# diamonds  
diamonds = sns.load_dataset("diamonds")

# lmplot  (clarity    )
p = sns.lmplot(data=diamonds.sample(5000),  # 5000  
            x="carat", 
            y="price", 
            hue="cut",  # cut   
            col="clarity",  # clarity   
            col_wrap=4)  #   4  

#     
p.set_titles(size=15)
p.set_xlabels("Carat", size=15)
p.set_ylabels("Price", size=15)

#  
plt.show()
```

```python
color_idx = (diamonds["color"] == 'D') | (diamonds["color"] == 'E') | (diamonds["color"] == 'F') 
cut_idx = (diamonds["cut"] == "Ideal") | (diamonds["cut"] == "Good") | (diamonds["cut"] == "Fair")
print(color_idx)
final_idx = color_idx & cut_idx
print(final_idx)

diamonds_df = diamonds[final_idx].copy()
diamonds_df["color"] = diamonds_df["color"].cat.remove_unused_categories()  
diamonds_df["cut"] = diamonds_df["cut"].cat.remove_unused_categories()  

# diamonds_df.shape
```

```python
# # diamonds  
# diamonds_df = sns.load_dataset("diamonds")

#   
fig, ax = plt.subplots(figsize=(6, 6))

# scatterplot 
sns.scatterplot(data=diamonds_df.sample(n=1000),  # 1000  
                x="carat", 
                y="price", 
                hue="color", 
                style="cut", 
                s=100)  

#  
plt.show()
```
