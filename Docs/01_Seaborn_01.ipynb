```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family = "gulim")
plt.rc('axes', unicode_minus = False)
```

```python
titanic = sns.load_dataset("titanic")
print("titanic = \n", titanic.head())
print(titanic.info())
```

```python
## missing
print("missing value = ")
# print(titanic.isnull().sum())

titanic_df = titanic.loc[:, :"fare"]
titanic_df.info()

titanic_df.dropna(axis = 0, inplace= True)
titanic_df.info()
```

```python
##
titanic_df.head()
```

```python
## sns.regplot
fig, ax = plt.subplots(figsize = (5,5))
sns.regplot(titanic_df, x = "age", y= "fare")
plt.show()
# fig.savefig("titanic_reg.png")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 예시 Titanic 데이터셋 불러오기 (사용하는 titanic_df가 정의되어 있는지 확인하세요)
# titanic_df = sns.load_dataset("titanic")

# regplot 그리기
sns.regplot(data=titanic_df,  # 'titanic_df'는 미리 정의된 데이터프레임이어야 합니다
            x="age", y="fare",
            fit_reg=True,
            color="r",
            marker="*",
            order=1,
            label="titanic data",
            scatter_kws={"fc": "b", "ec": "b", "s": 100, "alpha": 0.3},
            line_kws={"lw": 2, "ls": "--", "alpha": 0.8}  # 쉼표 추가
            )
plt.title("나이 VS 요금")
plt.xlabel("나이")
plt.ylabel("요금")
# 범례 추가
plt.legend(loc=0)

# 그래프 보여주기
plt.show()
```

```python
# diamonds 데이터셋 로드
diamonds = sns.load_dataset("diamonds")
diamonds_df = diamonds[diamonds["cut"] == "Ideal"]

# regplot 그리기
sns.regplot(data=diamonds_df.sample(n=1000),  # 1000개의 샘플 추출
            x="carat",
            y="price",
            order=2,
            scatter_kws={"fc": "b", "ec": "b", "s": 100, "alpha": 0.5},
            line_kws={"color": "red"}  # color= 수정
            )

# 제목 및 축 라벨 추가
plt.title("Carat VS Price")
plt.xlabel("Carat")
plt.ylabel("Price")

# 그래프 보여주기
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# diamonds 데이터셋 로드
diamonds = sns.load_dataset("diamonds")

# lmplot 그리기 (clarity에 따라 여러 패널로 나누기)
p = sns.lmplot(data=diamonds.sample(5000),  # 5000개의 샘플 사용
            x="carat", 
            y="price", 
            hue="cut",  # cut에 따라 색상 구분
            col="clarity",  # clarity에 따라 열로 나누기
            col_wrap=4)  # 한 줄에 4개의 그래프 배치

# 제목 및 라벨 크기 설정
p.set_titles(size=15)
p.set_xlabels("Carat", size=15)
p.set_ylabels("Price", size=15)

# 그래프 보여주기
plt.show()
```

```python
color_idx = (diamonds["color"] == 'D') | (diamonds["color"] == 'E') | (diamonds["color"] == 'F') 
cut_idx = (diamonds["cut"] == "Ideal") | (diamonds["cut"] == "Good") | (diamonds["cut"] == "Fair")
print(color_idx)
final_idx = color_idx & cut_idx
print(final_idx)

diamonds_df = diamonds[final_idx].copy()
diamonds_df["color"] = diamonds_df["color"].cat.remove_unused_categories()  
diamonds_df["cut"] = diamonds_df["cut"].cat.remove_unused_categories()  

# diamonds_df.shape
```

```python
# # diamonds 데이터셋 로드
# diamonds_df = sns.load_dataset("diamonds")

# 피겨와 축 생성
fig, ax = plt.subplots(figsize=(6, 6))

# scatterplot 그리기
sns.scatterplot(data=diamonds_df.sample(n=1000),  # 1000개의 샘플 사용
                x="carat", 
                y="price", 
                hue="color", 
                style="cut", 
                s=100)  

# 그래프 보여주기
plt.show()
```
